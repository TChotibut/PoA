### Computations of Lyapunov Exponents, crucial for Figure 9 and 10, Price of Anarchy, NeurIPS 2020

```python
def BifurcationDiagram(b_min=0.0, b_max=0.5, b_delta=0.001, g=f_MWUe, x0=0.3, n_skip =1000, n_iterate =200):
    figure(figsize=(50,20))
    xlim((b_min,b_max))
    b_array = arange(b_min, b_max, b_delta)
    bs = []
    trajs = []
    for b in b_array:
        bs.extend([b]*n_iterate)
        trajs.extend(IterateList(x0, a, b, n_iterate , n_skip , g))
    scatter(bs, trajs, marker = '.', s=0.2)
    #plot the superstable point 1/2 \pm \sqrt(1/4-1/mu)
    #xstable = 0.5+numpy.sqrt(1./4.-1./eta)
    plot([b_min,b_max],[0.059041448155901566,0.059041448155901566],'g-')
    setp(axes().get_xticklabels(),fontsize=40)
    setp(axes().get_yticklabels(),fontsize=40)
    ylabel('x', fontsize=40)
    
def Get_Lyapunov_Exponents(x, a, b_min=0.1, b_max=0.5, b_delta=0.001, g=f_MWUe, x0=0.49, n_skip =200, n_iterate =200):
    b_array = arange(b_min, b_max, b_delta)
    Lyapunov_exponents = []
    for b in b_array:
        Lyapunov_exponents.append(Lyapunov_Exponent(x, a, b, n_iterate, n_skip , g))
    return Lyapunov_exponents

def Plot_Lyapunov_exponents(Lyapunov_exponents, b_min, b_max, b_delta):
    figure(figsize=(50,20))
    xlim((b_min,b_max))
    b_array = arange(b_min, b_max, b_delta)
    plot(b_array, Lyapunov_exponents)
    plot([b_min,b_max],[0,0],'g-')
    setp(axes().get_xticklabels(),fontsize=40)
    setp(axes().get_yticklabels(),fontsize=40)
    ylabel('lyapunov_exponent', fontsize=40)
```
